```python
from dotenv import load_dotenv
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import Milvus
from langchain_openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter

load_dotenv()
```

```python
#loader = PyPDFLoader("../data/om_lfe_omz0012en.pdf", extract_images=False)
loader = PyPDFLoader("../data/om_pf2d5xx_omg0006en-f.pdf", extract_images=False)

pages = loader.load()
print(len(pages))
print(pages[0])
```

```python
text_splitter = CharacterTextSplitter(chunk_size=10, chunk_overlap=0)
docs = text_splitter.split_documents(pages)
```

```python
print(len(docs))
print(docs[0])
```

```python
embeddings = OpenAIEmbeddings()
```

```python
milvus = Milvus( embeddings, connection_args={"host": "localhost", "port": "19530"})
```

```python
milvus.add_documents(docs)
```
